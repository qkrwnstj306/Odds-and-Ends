```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Create a gradient image
gradient = np.linspace(0, 1, 256).reshape(1, -1)
gradient = np.vstack((gradient, gradient))

# Apply the JET colormap
jet_colormap = cv.applyColorMap((gradient * 255).astype(np.uint8), cv.COLORMAP_JET)
plt.figure(figsize=(20, 18))
# Display the colormap using matplotlib
plt.imshow(jet_colormap)
plt.title('JET Colormap')

# Create a custom color bar with reversed colors and values
cmap = LinearSegmentedColormap.from_list('custom_jet', jet_colormap[0, :, :] / 255.0, N=256)
bounds = np.linspace(0, 1, 256)
norm = plt.Normalize(0, 1)

# Reverse the ticks
ticks = np.linspace(1, 0, 11)

cb = plt.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap), ticks=ticks, boundaries=bounds)
#cb.set_label('Values')

# Set reversed ticks and labels
cb.set_ticks(1- ticks)
cb.set_ticklabels(['{:.1f}'.format(val) for val in ticks], fontsize=30)

# Invert the colorbar
cb.ax.invert_yaxis()

plt.show()
```
