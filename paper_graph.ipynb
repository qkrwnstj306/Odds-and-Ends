```python
import matplotlib.pyplot as plt
import numpy as np
def setPlotStyle():
    #plt.rcParams['font.family'] = font_name 
    plt.rcParams['font.size'] = 13
    plt.rcParams['legend.fontsize'] =  17
    plt.rcParams['lines.linewidth'] =  1
    plt.rcParams['lines.color'] = 'r'
    plt.rcParams['axes.grid'] = 1
    plt.rcParams['axes.xmargin'] = 0.1
    plt.rcParams['axes.ymargin'] = 0.1
    plt.rcParams['mathtext.fontset'] = "dejavuserif"
    #plt.rcParams['figure.dpi'] = 500
    #plt.rcParams['savefig.dpi'] = 500
    plt.style.use('seaborn-whitegrid')
    
if __name__=='__main__':
    x = [500, 750, 1000, 2000, 3000]
    setPlotStyle()
    fid_score = {
        'real_w/o_re' : [15.19692737,16.02800667,17.86418071,19.98986785,25.37206446],
        'mid_w/o_re' : [17.4898271,15.4490361,17.49949997,22.67660385,28.87257061],
        'anime_w/o_re' : [28.89221281,32.70074923,22.22745859,33.67670854,35.00397876],
        
        'real_with_re' : [13.00867763,14.04375702,15.80662697,22.64675188,24.65615416],
        'mid_with_re' : [15.56580252,12.22227925,15.92487737,16.2586085,25.53715513],
        'anime_with_re' : [29.95483429,27.88506878,20.71865743,25.80240536,33.13987263],
        
        'real_with_re2' : [13.00867763,15.08215774,16.06284473,21.82978627,25.36107982],
        'mid_with_re2' : [13.04703297,12.22227925,14.08785643,15.17014313,20.58421303],
        'anime_with_re2' : [30.27812876,30.60421826,21.40544762,25.63762231,24.53708599]
    }
    
    fid_name =  {
        '1' : 'realistic',
        '1-1' : 'real_w/o_re',
        '1-2' : 'real_with_re',
        '1-3' : 'real_with_re2',
        '2' : 'mid-journey',
        '2-1' : 'mid_w/o_re',
        '2-2' : 'mid_with_re',
        '2-3' : 'mid_with_re2',
        '3' : 'anime',
        '3-1' : 'anime_w/o_re',
        '3-2' : 'anime_with_re',
        '3-3' : 'anime_with_re2',
        
    }
    rows, cols = 1, 3 
    figure = plt.figure(figsize=(20,7))
    
    for i in range(1, rows * cols + 1):
        figure.add_subplot(rows, cols, i)
        mean_fid = (np.array(fid_score[fid_name[str(i)+'-2']]) + np.array(fid_score[fid_name[str(i)+'-3']])) / 2
        
        square_value = np.power(fid_score[fid_name[str(i)+'-2']],2) + np.power(fid_score[fid_name[str(i)+'-3']],2)
        e_square_value = square_value /2
        var = e_square_value - np.power(mean_fid,2)
        
        std = np.sqrt(var)
        plt.fill_between(x, mean_fid - std, mean_fid + std, alpha= 0.1, color = 'r', ls='--', label = 'std range')
        plt.title(fid_name[str(i)])
        
        plt.plot(x, fid_score[fid_name[str(i)+'-1']], '^-', label='W/O auxiliary')

        plt.plot(x, mean_fid, 'rD-', label = 'W/ auxiliary')
        plt.xticks([500, 750, 1000, 2000, 3000], fontsize=10)
        plt.legend(loc='best')
        plt.xlabel('train step')
        plt.ylabel('FID score')
    plt.show()
        
  
```
